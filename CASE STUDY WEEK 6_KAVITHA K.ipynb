```python
#Average scores in Math, Writing and Reading based on test preparation course and gender
# for some basic operations
import numpy as np
import pandas as pd
import math
data = pd.read_csv('StudentsPerformance.csv')
av_column = data.mean(axis=0)
print('Average scores:\n',av_column )
data[['test preparation course',
      'gender',
      'math score',
      'writing score',
      'reading score']].groupby(['test preparation course','gender']).agg('mean')

```

The average scores are given above.According to gender and test preparation course, females scored the most on average score in maths who completed test preparation course and males scored the most of average in maths who didn't appear in test preparation,female scored the most on average score in writing with no preparation course,female scored most on average in reading with no preparation course and males with preparation.

```python
#the scoring variation for math, reading and writing based on Gender and Test preparation course
# for some basic operations
import numpy as np
import pandas as pd
import math
data = pd.read_csv('StudentsPerformance.csv')
variance=data['math score'].var()
variance1=data['writing score'].var()
variance2=data['reading score'].var()
print('\n variance:\n',variance,variance1,variance2)
data[['test preparation course',
      'gender',
      'math score',
      'writing score',
      'reading score']].groupby(['test preparation course','gender']).agg('var')

```

High variance means that the data points  spread out from the mean, and from one another.Math score is highly heterogenous for both gender as well as for test preparation course.
